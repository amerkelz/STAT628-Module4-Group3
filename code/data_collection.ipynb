```python
top_50_shows = [x.replace('https://open.spotify.com/show/','') for x in show_urls] #ids for the top 50 shows in the US on spotify, according to https://podcastcharts.byspotify.com/
```

```python
def get_episode_ids(show_id):
    #return list of up to 200 episode ids for a given show
    result_lst = []
    offset_idx = 0
    call_ct = 0
    while True:
        api_results = sp.show_episodes(show_id, limit=50,offset=offset_idx)
        result_ct = len(api_results['items'])
        call_lst = [api_results['items'][i]['id'] for i in range(result_ct) if (api_results['items'][i] is not None) ]
        result_lst = result_lst + call_lst
        offset_idx = result_ct - 1
        time.sleep(5)
        call_ct = call_ct + 1
        if (result_ct < 50) or (len(result_lst) >= 300) or (call_ct > 6):
            break
    print('retrieved episode ids for show id', show_id)
    return result_lst
```

```python
batch_num = 0
start_idx = 0
end_idx = 45
total_batches = len(episode_lst) // 45 + (len(episode_lst) % 45 > 0)
result_lst = []
while True:
    api_param = episode_lst[start_idx:end_idx]
    
    if len(api_param) == 0: 
        print('no ids in batch') #just in case batch counting fails, avoid making empty api call
        break
        
    api_results = sp.episodes(episode_lst[start_idx:end_idx])['episodes']
    result_lst = result_lst + api_results
    
    batch_num = batch_num + 1
    start_idx = start_idx + 45
    end_idx = end_idx + 45
    
    time.sleep(5)
    print('completed episode batch',batch_num)
    print('result_lst length:', len(result_lst))
    if batch_num == total_batches:
        break
```

```python
episode_data_lst = []
#structure results as a dataframe:
for idx, item in enumerate(result_lst):
    key_data = {'episode_name': item['name'], 'podcast_name': item['show']['name'], 'publisher': item['show']['publisher'], 'release_date': item['release_date'], 'episode_description':item['description']}
    episode_data_lst.append(key_data)

episode_data = pd.DataFrame(episode_data_lst)
episode_data.tail()
episode_data.to_csv('episode_data.csv',index=False)
print('episode data saved to csv!')
```

### Description Cleanup

```python
episode_data = pd.read_csv('episode_data.csv')

## omit URLs from description field, quotation marks, capitalizations, etc.
def remove_urls(desc):
    #based on code generated from ChatGPT
    url_pattern = r'https?://\S+|www\.\S+'
    clean_desc = re.sub(url_pattern,'',str(desc))
    return clean_desc

def remove_punc(desc):
    for punc in string.punctuation:
        desc = desc.replace(punc, ' ')
    return desc
    
def remove_ad_clause(desc):
    #omits common ad phrase from some podcast descriptions: 'learn more about your ad choices'
    ad_clause1 = r'learn more about your ad choices'
    ad_clause2 = r'[Uu]se code'
    clean_desc = re.sub(ad_clause1,'',str(desc))
    clean_desc = re.sub(ad_clause2,'',str(clean_desc))
    clean_desc = clean_desc.replace('learn more about your ad choices',' ')
    return clean_desc

episode_data['clean_description'] = [remove_punc(remove_urls(episode_data.iloc[i,]['episode_description'])).lower() for i in range(len(episode_data['episode_description']))]
episode_data.to_csv('episode_data.csv')
```
